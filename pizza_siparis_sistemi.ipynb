```python
#2 Gerekli Kitaplıkları İçe Aktarma
import csv
import datetime


#3  "Menu.txt" dosyasını oluşturma

with open('Menu.txt', 'w') as menu:
  menu.write('*Lütfen bir pizza tabani seçiniz: \n')
  menu.write('1: Klasik 12.5 ₺\n')
  menu.write('2: Margarita 15 ₺\n')
  menu.write('3: TurkPizza 18 ₺ \n')
  menu.write('4: SadePizza 17.5 ₺ \n')
  menu.write('* ve seçeceğiniz sos: \n')
  menu.write('11: Zeytin 5.6 ₺ \n')
  menu.write('12: Mantarlar 7.5 ₺ \n')
  menu.write('13: Keçi Peyniri 7 ₺\n')
  menu.write('14: Et 15.5 ₺\n')
  menu.write('15: Soğan 4.7 ₺\n')
  menu.write('16: Mısır 7.2 ₺ \n')
  menu.write('* Teşekkürler \n')
  
#4 Ust sinif oluşturma :  "pizza"

class Pizza:
  def __init__(self, aciklama, maliyet):
    self.__aciklama = aciklama
    self.__maliyet = maliyet

  def get_description(self):
    return self.__aciklama

  def get_cost(self):
    return self.__maliyet
    

#5 Alt sinif oluşturma : "pizza"

class Klasik(Pizza):
  def __init__(self):
    aciklama = "Klasik Pizza "
    maliyet = 12.5
    Pizza.__init__(self, aciklama, maliyet)

class Margarita(Pizza):
  def __init__(self):
    aciklama = "Margarita Pizza "
    maliyet = 15
    Pizza.__init__(self, aciklama, maliyet)

class TurkPizza(Pizza):
  def __init__(self):
    aciklama = "Turk Pizza "
    maliyet = 18
    Pizza.__init__(self, aciklama, maliyet)

class SadePizza(Pizza):
  def __init__(self):
    aciklama = "Sade Pizza "
    maliyet = 17.5
    Pizza.__init__(self, aciklama, maliyet)

#6 Decorator ust sinifinin tanimlanmasi .Bir Decorator sınıfı oluşturun.
# Decorator, burada tüm sos sınıflarının süper sınıfı olarak adlandırılır.

class Decorator(Pizza):
  def __init__(self, component, aciklama, maliyet):
    self.component = component
    self.__aciklama = aciklama # kapsulleme
    self.__maliyet = maliyet
    Pizza.__init__(self, aciklama, maliyet) # pizza sınıfından miras alma islemi

  def get_cost(self):
    return Pizza.get_cost(self) 

  def get_description(self):
    return Pizza.get_description(self) 

#7 sos malzemelerinin tanimlanması
class Zeytin(Decorator):
  def __init__(self,component):
    aciklama = "Zeytin Soslu"
    maliyet = 5.6
    Decorator.__init__(self, component, aciklama, maliyet)

class Mantarlar(Decorator):
  def __init__(self,component):
    aciklama = "Mantar Soslu "
    maliyet = 7.5
    Decorator.__init__(self, component, aciklama, maliyet)

class KeciPeyniri(Decorator):
  def __init__(self,component):
    aciklama = "Keci Peyniri Soslu "
    maliyet = 7
    Decorator.__init__(self, component, aciklama, maliyet)

class Et(Decorator):
  def __init__(self,component):
    aciklama = "Et Soslu "
    maliyet = 15.5
    Decorator.__init__(self, component, aciklama, maliyet)

class Sogan(Decorator):
  def __init__(self,component):
    aciklama = "Sogan Soslu "
    maliyet = 4.7
    Decorator.__init__(self, component, aciklama, maliyet)

class Misir(Decorator):
  def __init__(self,component):
    aciklama= "Misir Soslu "
    maliyet = 7.2
    Decorator.__init__(self, component, aciklama, maliyet)


def Main():
    hosgeldiniz =''' \t \t \t
 ██░ ██  ▒█████    ██████   ▄████ ▓█████  ██▓    ▓█████▄  ██▓ ███▄    █  ██▓▒███████▒
▓██░ ██▒▒██▒  ██▒▒██    ▒  ██▒ ▀█▒▓█   ▀ ▓██▒    ▒██▀ ██▌▓██▒ ██ ▀█   █ ▓██▒▒ ▒ ▒ ▄▀░
▒██▀▀██░▒██░  ██▒░ ▓██▄   ▒██░▄▄▄░▒███   ▒██░    ░██   █▌▒██▒▓██  ▀█ ██▒▒██▒░ ▒ ▄▀▒░ 
░▓█ ░██ ▒██   ██░  ▒   ██▒░▓█  ██▓▒▓█  ▄ ▒██░    ░▓█▄   ▌░██░▓██▒  ▐▌██▒░██░  ▄▀▒   ░
░▓█▒░██▓░ ████▓▒░▒██████▒▒░▒▓███▀▒░▒████▒░██████▒░▒████▓ ░██░▒██░   ▓██░░██░▒███████▒
 ▒ ░░▒░▒░ ▒░▒░▒░ ▒ ▒▓▒ ▒ ░ ░▒   ▒ ░░ ▒░ ░░ ▒░▓  ░ ▒▒▓  ▒ ░▓  ░ ▒░   ▒ ▒ ░▓  ░▒▒ ▓░▒░▒
 ▒ ░▒░ ░  ░ ▒ ▒░ ░ ░▒  ░ ░  ░   ░  ░ ░  ░░ ░ ▒  ░ ░ ▒  ▒  ▒ ░░ ░░   ░ ▒░ ▒ ░░░▒ ▒ ░ ▒
 ░  ░░ ░░ ░ ░ ▒  ░  ░  ░  ░ ░   ░    ░     ░ ░    ░ ░  ░  ▒ ░   ░   ░ ░  ▒ ░░ ░ ░ ░ ░
 ░  ░  ░    ░ ░        ░        ░    ░  ░    ░  ░   ░     ░           ░  ░    ░ ░    
                                                  ░                         ░        '''
    

    print(hosgeldiniz)
    menu = open("Menu.txt")
    print(menu.read())

#8 Kullanici tarafindan pizza tabani secilmesi

    pizza_tabani = input("Lutfen bir pizza tabani seciniz (1-4) : ")
    while pizza_tabani not in ["1", "2", "3", "4"]:
      print(f"Suanda {pizza_tabani} pizzamız yok.")
      pizza_tabani = input("Lutfen menumuzden bir pizza numarası seçiniz (1-4): ")

    if pizza_tabani == "1":
      pizza_siparis = Klasik()
    elif pizza_tabani == "2":
      pizza_siparis = Margarita()
    elif pizza_tabani == "3":
      pizza_siparis = TurkPizza()
    elif pizza_tabani == "4":
      pizza_siparis = SadePizza()
    

        
#9 Kullanici tarafindan sos secilmesi
    sos_tipi = input("Lutfen bir sos seciniz (11-16): ")
    while sos_tipi not in ["11", "12", "13", "14", "15", "16"]:
      print(f"Suanda {sos_tipi} sosumuz yok .")
      sos_tipi = input("Lutfen menumuzden bir pizza sosu numarası seçiniz (11-16): ")
    
    if sos_tipi == "11":
      sos = Zeytin(Pizza)
    elif sos_tipi == "12":
      sos = Mantarlar(Pizza)
    elif sos_tipi == "13":
      sos = KeciPeyniri(Pizza)
    elif sos_tipi == "14":
      sos = Et(Pizza)
    elif sos_tipi == "15":
      sos = Sogan(Pizza)
    elif sos_tipi == "16":
      sos = Misir(Pizza)
    
#10 siparis bilgilerinin alinmasi
    siparis_toplami = pizza_siparis.get_cost()+sos.get_cost()
    isim = input("Adinizi ve soyadinizi giriniz: ")
    kimlik_no = input("Kimlik numaranizi giriniz: ")
    kart_no = input("Kredi karti numaranizi giriniz: ")
    kart_sifre = input("Kredi karti sifrenizi giriniz: ")
    

    suan = datetime.datetime.now()
    tarih = datetime.datetime.strftime(suan, '%c')

#11 siparislerin kaydedilmesi
    with open ('Orders_Database.csv', mode='a') as siparis_kayitlari:
      yeni_kayit =csv.writer(siparis_kayitlari, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
      yeni_kayit.writerow([pizza_siparis.__class__.__name__, isim, kimlik_no, kart_no, kart_sifre,sos.get_description() + pizza_siparis.get_description(), tarih, siparis_toplami])
        
    print("\n")
    
    print(f"Sayin {isim} , {siparis_toplami} ₺ tutarindaki {sos.get_description()} {pizza_siparis.get_description()} siparisiniz icin; ")
    print(f" {tarih} tarihinde {kart_no} numarali kredi kartinizdan {siparis_toplami} ₺ odeme gerceklestirilmistir . Bilgilerinize sunariz .")

Main()

print("\n")


tesekkur = '''
████████╗███████╗███████╗███████╗██╗  ██╗██╗  ██╗██╗   ██╗██████╗     ███████╗██████╗ ███████╗██████╗ ██╗███████╗
╚══██╔══╝██╔════╝██╔════╝██╔════╝██║ ██╔╝██║ ██╔╝██║   ██║██╔══██╗    ██╔════╝██╔══██╗██╔════╝██╔══██╗██║╚══███╔╝
   ██║   █████╗  ███████╗█████╗  █████╔╝ █████╔╝ ██║   ██║██████╔╝    █████╗  ██║  ██║█████╗  ██████╔╝██║  ███╔╝ 
   ██║   ██╔══╝  ╚════██║██╔══╝  ██╔═██╗ ██╔═██╗ ██║   ██║██╔══██╗    ██╔══╝  ██║  ██║██╔══╝  ██╔══██╗██║ ███╔╝  
   ██║   ███████╗███████║███████╗██║  ██╗██║  ██╗╚██████╔╝██║  ██║    ███████╗██████╔╝███████╗██║  ██║██║███████╗
   ╚═╝   ╚══════╝╚══════╝╚══════╝╚═╝  ╚═╝╚═╝  ╚═╝ ╚═════╝ ╚═╝  ╚═╝    ╚══════╝╚═════╝ ╚══════╝╚═╝  ╚═╝╚═╝╚══════╝
                                                                                                                 '''
print(tesekkur)

#12 siparis bilgilerinin tablosunun gösterilmesi

import pandas as pd
print()
print()
df = pd.read_csv("Orders_Database.csv")
df.columns=["Pizza Tabani","isim - soy isim","Kullanici Kimligi", "Kredi Karti No","Kredi Karti Sifresi ","Siparis Aciklamasi ","Siparis Zamani","Siparis Tutari"]
df



  
```
